# Import Libraries

```python
import numpy as np
import pandas as pd
```

## Create DataFrame

```python
customer_Df = pd.DataFrame({
    
    'Date':['2015-08-24 01:10:24', '2015-08-24 01:15:32', '2015-09-08 12:08:03'],
    'Customer':['A','B','A'],
    'Data Center':['X','Y','W'],
    'Duration':['125','15000','225']
})
customer_Df
```

```python
customer_Df.info()
```

```python
customer_Df.describe()
```

```python
cus_agg = customer_Df.groupby('Customer').agg(Total_Duration =('Duration','sum')).reset_index()
cus_agg
```

```python
customer_Df['new_date'] = pd.to_datetime(customer_Df['Date'])
customer_Df
```

```python
customer_Df['Only_date'] = customer_Df['new_date'].dt.date
customer_Df
```

```python
customer_Df['Only_time'] = customer_Df['new_date'].dt.time
customer_Df
```

```python
customer_Df['Day_name'] = customer_Df['new_date'].dt.day_name()
customer_Df
```

```python
customer_Df['Day_name'] = customer_Df['new_date'].dt.hour
customer_Df
```
