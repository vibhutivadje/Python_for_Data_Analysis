```python
train_df.groupby('Sex').Age.max()
```

```python
train_df.groupby('Sex').Age.agg(['min','max','median','count','median'])
```

```python
train_df.groupby('Sex').Age.agg(sex_max =('max'), sex_min=('min'))
```

```python
def categorize(x):
    m = x.mean()
    if m> 29:
        return True
    else:
        return False
```

```python
train_df.groupby('Sex').Age.agg(['min','max',categorize])
```

```python
train_df.groupby('Sex').Age.agg(['min','max',lambda x: True if x.mean()> 29 else False])
```

```python
train_df.groupby('Cabin').size()
```

```python
train_df.groupby('Cabin').filter(lambda x: len(x) >=4)
```
