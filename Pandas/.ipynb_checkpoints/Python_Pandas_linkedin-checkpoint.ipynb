<B>Create dataframe with random numbers shape(6,6)</B>

```python
df_obj = pd.DataFrame(np.random.rand(36).reshape((6,6)), 
                      index = ['row1','row2','row3','row4','row5','row6'],
                      columns = ['col1','col2','col3','col4','col5','col6'])
df_obj
```

<B>Retrieve values from (row2 & row5) AND (col5 & col2)</B>

```python
df_obj.loc[['row2','row5'], ['col5','col2']]
```

<B>Data Slicing</B>

```python
df_obj.loc['row2':'row5']
```

```python
#this will only compare 
df_obj < .2
```

```python
#to print filtered objects
trun_df = df_obj[df_obj < .2]
trun_df
```

```python
df_obj.loc[['row1'],['col1']] = 8
df_obj
```
