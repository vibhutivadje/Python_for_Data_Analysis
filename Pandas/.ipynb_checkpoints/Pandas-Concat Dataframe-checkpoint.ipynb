Note: Concat will add two dataframes and merge will add dataframes on values

```python
import pandas as pd
```

```python
india_weather= pd.DataFrame({
    "city":["mumbai","pune","delhi"],
    "temperature":[32,45,69],
    "humidity":[80,70,60]
})
india_weather
```

```python
us_weather= pd.DataFrame({
    "city":["denver","seattle","california"],
    "temperature":[35,15,89],
    "humidity":[32,45,63]
})
us_weather
```

```python
weather_df = pd.concat([india_weather,us_weather])
weather_df
```

```python
weather_df = pd.concat([india_weather,us_weather],ignore_index = True)
weather_df
```

```python
weather_df = pd.concat([india_weather,us_weather], keys=["india",'us'])
weather_df
```

```python
weather_df.loc['india']
```

```python
weather_df.loc['us']
```

```python
temp_df= pd.DataFrame({
    "city":["mumbai","pune","delhi"],
    "temperature":[32,45,69]
})
temp_df
```

```python
humidity_df = pd.DataFrame({
    "city":["mumbai","pune","delhi"],
    "humidity":[80,70,60]
})
humidity_df
```

```python
weather_india = pd.concat([temp_df,humidity_df], axis = 1)
weather_india
```

```python
#merge series and dataframe
s = pd.Series(["Humid","Dry","Rain"], name = "event")
s
```

```python
f_df = pd.concat([weather_india,s], axis = 1)
f_df
```
