```python
#writing converter functions:

def convert_people_cell(cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell

new_df = pd.read_excel('/Users/vibhutivadje/Documents/Git_Repos/Python_for_Data_Analysis/Pandas/Datasets/stock_data.xlsx',
                       'Sheet1',
                       converters = {
                           'people': convert_people_cell,
                           'eps': convert_eps_cell
                       })

new_df
    
```
